<h2>Linear Plot</h2>

<h3> Line charts are used to represent the relation between two data X and Y on a different axis.</h3>

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])

plt.plot([1,4,5],[10,20,30])    # [1,4,5] are x-axis values and [10,20,30] are y-axis values.

# If we don't want this to print '[<matplotlib.lines.Line2D at 0x26fe272dc90>]' like in previous output we use plt.show() 
plt.show()
```

```python
# To change the color of the line

plt.plot([1,4,5],[10,20,30],color="yellow")
```

```python
# to manage/give x-axis and y-axis values in the graph

price = [20000,30000,40000,10000]
year = [2020,2021,2022,2023]

labels = [2020,2021,2022,2023]
labs = [10000,20000,30000,40000]

plt.xticks(year, labels)
plt.yticks(price,labs)

plt.plot(year,price,marker="o")
plt.show()
```

```python
# To create a sin graph plot

import numpy as np
x = np.linspace(0,10,100)    # divides the distance between 0&10 into 100 equal parts (1 part = 0.1)

x1, y1 = [0, 10], [0, 0]   # to draw a straight line in the graph x1 is the points [0,10] is from 0 to 10 ,same with y1. 
plt.plot(x1, y1, marker = 'o', color='black')

plt.plot(x,np.sin(x))
```

```python
import numpy as np
x = np.linspace(0,10,100)    # divides the distance between 0&10 into 100 equal parts (1 part = 0.1)

x2,y2 = [0,10],[0,0]
plt.plot(x2,y2,marker='o', color="blue")

plt.plot(x,np.cos(x))
```

```python
# to draw many line grpahs in one graphs, just use plt.plot()

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

x1,y1 = [0,10],[0,0]
plt.plot(x1,y1, marker='*',ms='10')   # ms is marker size.

plt.show()
```

```python
# To change the line style in graph

plt.plot(x,np.sin(x),linestyle="solid") # or linestyle="-"
plt.plot(x,np.sin(x+2),linestyle="dashed") # or linestyle="--"
plt.plot(x,np.sin(x+4),linestyle="dashdot") # or linestyle="-."
plt.plot(x,np.sin(x+6),linestyle="dotted") # or linestyle."
plt.show()
```

```python
# Shortcut for color and linestyle.

plt.plot(x,np.sin(x),"-g") # or linestyle="-"
plt.plot(x,np.sin(x+2),"--k") # or linestyle="--"
plt.plot(x,np.sin(x+4),"-.b") # or linestyle="-."
plt.plot(x,np.sin(x+6),".y") # or linestyle = "."
plt.show()
```

```python
# to manage x-axis values and y-axis values

plt.xlim(0,5)
plt.ylim(-4,4)
plt.show()
```

```python
plt.plot(x,np.sin(x))

# to give names to the axes and grpah

plt.xlabel("this is x-axis")
plt.ylabel("this is y-axis")
plt.title("this is sin graph")

plt.show()
```

```python
plt.plot(x,np.sin(x),label="sin")
plt.plot(x,np.cos(x),label="cos")
plt.legend()           # the box(sin and cos) is given by this legend method.

plt.show()
```

<h2>Scatter plot</h2>

<h3>We use scatterplot points to show the relation between two sets of data.</h3>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10)
y = np.cos(x)

plt.plot(x,y,"o")    # "o" is the marker we should not write marker keyword here.
plt.show()
```

<h2>Histogram Plot</h2>

<h3>Histogram is used to display statistical information which uses rectangle to show frequency of data items in successive numerical intervals of equal sides.</h3>

```python
data1 = [10,30,20,25,35,31,32,33]

plt.hist(data1,bins=[0,10,20,30,40,50])
plt.ylabel('frequency')

plt.show()


# y-axis is the frequency.It shows no. of numbers occured in the range.like from above ex. from 0-10 no num occured. from 10-20
# one number occured so freq. is 1. From 20-30 2 nums occured so freq. is 2. From 30-40 5 nums occured so freq. is 5.
```

```python
marks = [10,20,80,30,90,40,50,40,50]
ranges = [0,30,60,90]

plt.hist(marks,ranges,color="lavender")
plt.xticks([0,30,60,90])   # the plot indicates that form 0 to 30, 2 students are there. From 30 to 60, 5 students are there. From 60 to 90, 2 students are there. 
plt.title('the frequency of students and their marks')
plt.xlabel('marks range')
plt.ylabel('number of students')

plt.show()
```

<h2>Bar Plot</h2>

<h3>Bar plot represents the category of data with rectangular bars with lengths and heights proportional to values which they represent.</h3>

```python
import matplotlib.pyplot as plt

colors = ["black","red","blue","green","pink","yellow"]
people = [70,40,10,20,60,50]

plt.bar(colors,people)

plt.show()
```

```python
import matplotlib.pyplot as plt

colors = ["black","red","blue","green","pink","yellow"]
people = [70,40,10,20,60,50]

plt.bar(colors,people)

plt.xticks(rotation='vertical')   # changes the x-labels to vertical position.

plt.show()
```

```python
plt.barh(colors,people,color="pink")     # barh is used to plot bar graph horizontally.
```

```python
# Multiple bar charts in a plot

import numpy as np

subject = ['c program','maths','python','physics','aiml']   # x-axis labels

# These are the branches and frequencies of the each label in x-axis.
cse = [100,40,30,80,40]
it = [80,40,40,10,10]       
ece = [30,50,10,90,90]
eee = [10,70,60,50,60]

wdth = 0.2                      # this is the width of the each bar.


# this is used to increase the size of the plot.
plt.figure(figsize=(10,5))


# these are the points used to plot multiple branches to single subject  (Locations to plot)
cse_num = np.arange(5)
it_num = [i+wdth for i in cse_num]
ece_num = [i+wdth for i in it_num]
eee_num = [i+wdth for i in ece_num]

# this is converting the python list it_num to numpy array.
new_it = np.array(it_num)

plt.bar(cse_num,cse,color='blue',width=wdth,label="CSE")
plt.bar(it_num,it,color='red',width=wdth,label="IT")
plt.bar(ece_num,ece,color='black',width=wdth,label="ECE")
plt.bar(eee_num,eee,color='orange',width=wdth,label="EEE")

# this is used to change the x-labels to subject names again.
plt.xticks(new_it + 0.1,subject)
plt.title('Marks Comparision')
plt.xlabel('S u b j e c t s')
plt.ylabel(' Average Marks')

plt.legend()
plt.show()
```

<h2>Pie Chart</h2>

<h3>Pie Chart is a circular statistical plot that can display only one series of data.</h3>

```python
import matplotlib.pyplot as plt

activity = ['Dancing','Singing','Learning','Watching']
time = [30,40,20,10]

plt.figure(figsize=(8,10))

plt.pie(time,labels=activity)
plt.title('Activities')

plt.legend(loc='lower left',title='Activites')

plt.show()
```

```python
import matplotlib.pyplot as plt

activity = ['Dancing','Singing','Learning','Watching']
time = [30,50,20,10]

plt.figure(figsize=(6,10))

plt.pie(time,labels=activity,explode=[0.1,0,0,0.2],shadow=True,colors=['green','purple'],autopct="%2.1f%%")          
# explode used to extract a part from the pie chart.
# autopct gives the perentage of the each part in the the pie circle. %2.1f%% is syntax. 2 represents the no. of digits to be 
# displayed before decimal. and 1 represents no. of digits to be displayed after decimal.

plt.title('Activities')

plt.show()
```

```python
## Subplots

import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,20,30,40,50]

x1 = np.linspace(0,10,100)
y1 = np.sin(x1)

y2 = np.cos(x1)

y3 = np.tan(x1)

## plt.subplot(total number of rows , total number of columns , index number)    #index number starts from 1 not 0.

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('line graph')

plt.subplot(2,2,2)
plt.plot(x1,y1)
plt.title('Sin graph')

plt.subplot(2,2,3)
plt.plot(x1,y2)
plt.title('Cos graph')

plt.subplot(2,2,4)
plt.plot(x1,y3)
plt.title('Tan graph')

plt.tight_layout()                     ## It is used to give spacing betweeen the rows instead of overlapping.

plt.show()
```
