```python
import seaborn as sns
import pandas as pd
```

<h3>Joint Probability</h3>

```python
df = sns.load_dataset('titanic')
df
```

```python
pd.crosstab(df['survived'],df['pclass'])
```

```python
pd.crosstab(df['survived'],df['pclass'],normalize=True)
```

<h4>0.089787, 0.108866,... are joint probabilities</h4>

<h3>Marginal probability</h3>

```python
pd.crosstab(df['survived'],df['pclass'],normalize=True,margins=True)
```

<h4>the All column and row are the marginal probablities.<br><br>Whereas the normal row and column values except All row and column are joint probabilities.</h4>

<h3 style="color:red">Conditional Probability</h3>

<h4>It is the measure of probability of an event occuring given that another event has already occured. given by P(A|B) = P(A intersection B)/P(B)</h4>

```python
pd.crosstab(df['survived'],df['pclass'],normalize="columns")
```

<h4>The above results indicate that if a person is from pclass x then the probability of him dying/surviving is corresponding value.<br><br>Ex. if the person is from pclass 1 then probability of dying is 0.37037 and surviving is 0.62963. same for all pclasses.</h4>

```python
pd.crosstab(df['survived'],df['pclass'],normalize="index")
```

<h4>Now when "index" is given in normalize, it states that if the person is dying, then the probability of him being from pclass 1 is 0.145719 and from pclass 2 is 0.176685.<br><br> And if the person is surviving then the probability of him being from pclass 1 is 0.397661. and so on.</h4>

<h3 style="color:red">Baye's Theorem</h3>

<h3> P(A|B) = (P(B|A) . P(A)) / P(B)</h3>
