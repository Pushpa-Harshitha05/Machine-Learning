<h2>(i) Scatterplot - (axis level function)</h2>

```python
import numpy as np
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt

tip = sns.load_dataset('tips')
tip
```

```python
total_bill_30 = tip['total_bill'].head(30)
tips_30 = tip['tip'].head(30)

sns.scatterplot(data=tip,x=total_bill_30,y=tips_30)
```

<h2>scatterplot is axis level plot. if we want to plot it as figure level plot we use relplot, same with lineplot</h2>

```python
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

dataset = sns.load_dataset('tips')

fig, ax = plt.subplots(figsize=(10, 5))

sns.scatterplot(data=dataset,x='total_bill',y='tip',hue='day',style='time',size='size')    ## hue divides the colour of the dots based on categories in the column.
   ## style also divides the shape based on the category.
   ## size changes the size of the dots based on the value.
```

<h2>(ii) Line plot - axis level function</h2>

```python
import plotly.express as px

gap = px.data.gapminder()
gap
```

```python
country = gap[gap['country'] == 'India']

sns.lineplot(data=country,x='year',y='lifeExp')      ## axis level plot  

sns.relplot(data=country,x='year',y='lifeExp',kind='line')     ## figure level plot
```

```python
df = gap[gap['country'].isin(['India','China','Germany'])]
df

sns.relplot(data=df,x='year',y='lifeExp',hue='country',kind='line',style='continent',size='continent')    # hue divides the line based on category.
```

<h2>Facet plot - It is like subplots with same plot.works with all types of figure level funtions, does not work with axis level functions.</h2>

```python
sns.relplot(data=tip,x='total_bill',y='tip',kind='scatter',row='sex',col='smoker')
```

<h1>2. Distirbution Plots</h1>

<h2>(i) Histogram plot - axis level function</h2>

```python
import seaborn as sns
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips')

sns.histplot(data=tips,x='total_bill')
```

<h2>Since the histogram is axis level function, so we use displot to plot figure level function of distribution plots, same with kde plot and rug plot.</h2>

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# another dataset example

titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,x='age',kind="hist",col='sex')
```

<h2>(ii) kde plot(kernel density estimation) - axis level function</h2>

```python
x1, y1 = [15.5,15.5], [0.00,0.053]
plt.plot(x1,y1,marker='o')

sns.kdeplot(data=tips,x='total_bill')

## kdeplot tells the probability like getting total_bill 15.5 has the highest probability 0.053%
```

```python
## axis level funtion
sns.kdeplot(data=tips,x='total_bill',hue='sex',fill=True)

## figure level function
sns.displot(data=tips,x='total_bill',hue='sex',fill=True,kind='kde')
```

<h2>(iii) Rug Plot - axis level function</h2>

<h3>Plot marginal distributions by drawing ticks on the x-axis</h3>

```python
plt.figure(figsize=(10,5))

sns.rugplot(data=tips,x='total_bill')
```

<h2>Bivariate Histogram and Bivariate kde plot</h2>

```python
# bivariate histplot (2-D plot)

sns.histplot(data=tips,x='total_bill',y='tip')
# (or)
#sns.displot(data=tips,x='total_bill',y='tip',kind='hist',height=10)   #to increae the size of figure-level axis we use height.

## dark colour box indicates that there are more people in that region. in this graph it indicates that there are more people 
## lies in the region where the total_bill is aprox. 15 and tip is aprox. 2.5
```

```python
## bivariate kdeplot

sns.kdeplot(data=tips,x='total_bill',y='tip')
# (or)
# sns.displot(data=tips,x='total_bill',y='tip',kind='kde')

## the region or the place in the graph where the lines are closer indicates more people (density).
```

<h1>3. Matrix plot</h1>

<h2>(i) Heat map - axis level function</h2>

```python
import plotly.express as px

gap = px.data.gapminder()
gap
```

```python
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')    # converting the long form data into wide format.

plt.figure(figsize=(10,10))
sns.heatmap(temp_df)
```

```python
## To display only asian countries

asia_df = gap[gap['continent'] == 'Europe']
asia_df = asia_df.pivot(index='country', columns='year', values='lifeExp')

plt.figure(figsize=(12, 10))

# Create the heatmap with annotations
sns.heatmap(asia_df, annot=True,square=True)
plt.show()
```

<h3>ClusterMap - plot a matrix dataset as a hierarchichally-clustered heatmap.</h3>
<h4>It actually makes similar columns together and seperates dissimilar columns.</h4>

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

<h4>The above clustermap uses machine learning algorithm (Agglomerative Hierarchical clustering algo) internally to tell the closeness of the given columns.</h4>
<h4>The above one tells that the values of petal_length and sepal_width are more closer, then petal_width and then sepal_length</h4>
<h4>And the numbers at the side indicates the closeness with each other.</h4>

<h1>4. Categorical Plots</h1>

<h2>(1). Categorical Scatter plots</h2>
<h3>We use categorical scaterr plots when we want to see the distr. of categorical data with respect to a numerical variable.It is bivariate analysis. </h3>

<h3>(i). Striplot - axis level function</h3>

```python
import seaborn as sns

tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
# normally scatterplot is plotted between 2 numerical columns. But stripplot is plotted between one categorical and one numerical columns.

sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)  # try jitter =True also.
```

<h3>To make axis level to figure level plots in categorical plots we use catplot.</h3>

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=True,hue='sex')
```

<h3>(ii) Swarm plot - axis level func.</h3>

```python
# it explains about the distribution of the data more understandibility.It is more usable for small datasets unlike stripplots where they are used for large,medium datasets.

sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')

## figure level:
#  sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

<h2>(i)Box plot - axis level func.</h2>

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
sns.boxplot(data=tips,y='total_bill')


## try these too
#sns.boxplot(data=tips,x='sex',y='total_bill')
#sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

<h2>3. Categorical Estimate plots -> for central tendency</h2>
<h3>The categorical estimate plots are mainly used for estimating some aggregate values like mean,standarad deviaiton,count</h3>

<h2>(i) Barplot - ais level func.</h2>
<h3>When there are multiple observations in each category, It also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars.</h3>

```python
sns.barplot(data=tips,x='sex',y='total_bill',estimator='mean') 

male = tips[tips['sex'] == 'Male']['total_bill'].mean()
print(male)

# the average total_bill for male is around 21 dollars and female is 18 dollars.
```

```python
import numpy as np

sns.barplot(data=tips,x='sex',y='total_bill',errorbar=None,estimator=np.std) 

#estimator default value = 'mean', we can change that.the other values are 'std','min','max','median'..

# errorbarshows the range within which the true value is expected to fall. Error bars help to visualize how precise or accurate your measurements are.
```

<h3>House Dataset</h3>

```python
from sklearn.datasets import fetch_california_housing

house = fetch_california_housing()
houses = pd.DataFrame(house.data,columns=house.feature_names)
houses
```

```python
correlation = houses.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={"size":8}, cmap='Blues')
plt.show()
```
