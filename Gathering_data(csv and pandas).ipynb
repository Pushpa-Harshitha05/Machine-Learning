<h1>Converters</h1><h3>used to change value of a col</h3>

```python
def rename(text):
  text = float(text)
  if text == 1021.3:
    return 0
  else:
    return 1
```

```python
pd.read_csv('https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Air%20Quality/real_2016_air.csv',converters={'SLP':rename})
```

<h1>Loading a huge dataset in chunks using chunksize=</h1>

```python
data = pd.read_csv('https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Air%20Quality/real_2016_air.csv',chunksize=100)

for chunks in data:
  print(chunks.shape)
```
