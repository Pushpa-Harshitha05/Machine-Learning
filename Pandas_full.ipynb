```python
import pandas as pd
import os

os.chdir("C:/Users/praka/OneDrive/Desktop")

movies = pd.read_csv("movies.csv")
movies
```

```python
movies.sort_values(by='imdb_rating',ascending=False,inplace=True)
movies
```

```python
# minimum imdb_rating for all movies
mini = movies['imdb_rating'].min()

r = movies.loc[movies['imdb_rating'] == mini, 'title']
r
```

```python
# maximum imdb_rating for all movies

maxi = movies['imdb_rating'].max()

r = movies.loc[movies['imdb_rating'] == maxi, 'title']
r
```

```python
# average imdb_rating for all movies
movies['imdb_rating'].mean()
```

```python
# minimum imdb_rating for only bollywood movies

bollywd = movies[movies['industry'] == 'Bollywood']
bollywd

mini = bollywd['imdb_rating'].min()

r = bollywd.loc[bollywd['imdb_rating'] == mini, 'title']
r
```

```python
# maximum imdb_rating for only bollywood movies

maxi = bollywd['imdb_rating'].max()

r = bollywd.loc[bollywd['imdb_rating'] == maxi, 'title']
r
```

```python
bollywd['imdb_rating'].mean()
```

```python
# minimum imdb_rating for only hollywood movies

hollywd = movies[movies['industry'] == 'Hollywood']
hollywd

mini = hollywd['imdb_rating'].min()

r = hollywd.loc[hollywd['imdb_rating'] == mini, 'title']
r
```
